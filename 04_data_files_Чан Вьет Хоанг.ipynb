3. По данным из файла `addres-book-q.xml` сформировать список словарей с телефонами каждого из людей. 

````python
from bs4 import BeautifulSoup
with open ("addres-book-q.xml","r") as f:
  lst=BeautifulSoup(f,"xml")
  list=[]
  for i  in lst.find_all('address'):
    name=i.find("name").get_text()
    for j in i.find_all('phone'):
      phone=j.get_text()
      list.append({name:phone})
list
````

1.5 Создайте `pd.DataFrame` `contributors`, имеющий столбцы `id`, `username` и `sex`.

````python
import pandas as pd
import json 
main=[]
user=[]
sx=[]
with open('contributors_sample.json')as f :
  beta=json.load(f)
  for i in range(len(beta)):
    main.append(beta[i]["id"])
  for i in range(len(beta)):
    user.append(beta[i]["username"])
  for i in range(len(beta)):
    sx.append(beta[i]["sex"])
  contributors= pd.DataFrame({'id':main,'username':user, 'sex': sx })
print(contributors)




````
